```python
plt.figure(figsize=(15,16))
sns.countplot(x="Sub-Category", data=data,orient='v')
```

```python
plt.figure(figsize=(15,12))
sns.countplot(x=data['City'], order=(data['City'].value_counts().head(50)).index)
plt.xticks(rotation=90)
```

```python
sns.countplot(x='State',data=data.head(100))
plt.xticks(rotation=90)
plt.show()
```

```python
sns.catplot('Ship Mode',data=data,hue='Segment',kind="count",height=5,aspect=1.5,palette="plasma")
```

```python
sns.pairplot(data,hue='Sub-Category')
```

```python
numerical_features=[features for features in data.columns if(data[features].dtype!='O')]
print(numerical_features)
print(len(numerical_features))
```

```python
sns.jointplot(x='Sales',y='Discount',data=data,kind='kde')
```

```python
sns.barplot(x=data['Sub-Category'], y=data['Profit'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.pie(data['Category'].value_counts(), labels=data['Category'].value_counts().index, startangle=180, radius=1)
plt.title('Main category of operation')
plt.show()
print (data['Category'].value_counts())
```

```python
plt.figure(figsize=(10,7))
data['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```
